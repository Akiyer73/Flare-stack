```python
ch0 =hlsimage[:,:,0]
blur = cv2.GaussianBlur(ch0,(3,3),0)
canny = cv2.Canny(ch0,200, 255)
```

```python
plt.imshow(canny)
points = []
minx = 200
miny = 200
maxx = 0
maxy = 0
for y in range(0, 240):
    x = 0
    for x in range(0,140):
        #find the minimum Y
        if canny[x,y]>0:
            point = [x,y]
            points.append(point)
            if minx > x:
                minx = x
            if miny > y:
                miny = y
            if maxx < x:
                maxx = x
            if maxy <y:
                maxy = y
            
        x += x
    y += y       
    
print(minx)
print(maxx)
print(miny)
print(maxy)

```
